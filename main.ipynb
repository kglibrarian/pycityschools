```python
# High-level snapshot of districts key metrics

#Total Schools
schools_total = schools_df["school_name"].count()
schools_total

#Total Students
students_total = students_df["Student ID"].count()
students_total

#Total Budget
budget_total = schools_df["budget"].sum()
budget_total

#Average Math Score
math_average_total = students_df["math_score"].mean()
math_average_total_round = round(math_average_total, 6)
math_average_total_round

#Average Reading Score
reading_average_total = students_df["reading_score"].mean()
reading_average_total_round = round(reading_average_total, 6)
reading_average_total_round

# % Passing Math
passing_math_total = (passing_math_df["Student ID"].count()/ students_total) * 100
passing_math_total_round = round(passing_math_total, 6)
passing_math_total_round

# % Passing Reading
passing_reading_total = (passing_reading_df["Student ID"].count()/ students_total) * 100
passing_reading_total_round = round(passing_reading_total, 6)
passing_reading_total_round

# Overall Passing Rate
overall_passing_rate = (passing_reading_total_round + passing_math_total_round)/2
overall_passing_rate

# Creating a summary DataFrame using above values
high_level_snapshot_df = pd.DataFrame({"Total Schools": [schools_total],
                           "Total Students": [students_total],
                           "Total Budget": [budget_total],
                           "Average Math Score": [math_average_total_round],
                           "Average Reading Score": [reading_average_total_round],
                           "% Passing Math": [passing_math_total_round],
                           "% Passing Reading": [passing_reading_total_round],
                           "Overall Passing Rate": [overall_passing_rate]})

high_level_snapshot_df
```

```python
#SCHOOL SUMMARY

#Average Reading Score by School
average_reading_score = students_df.groupby('school_name')['reading_score'].mean()
average_reading_score

avg_reading_score_df = pd.DataFrame({'school_name':average_reading_score.index, 'avg_reading_score':average_reading_score.values})
avg_reading_score_df

#Average Math Score by School
average_math_score = students_df.groupby('school_name')['math_score'].mean()
average_math_score

avg_math_score_df = pd.DataFrame({'school_name':average_math_score.index, 'avg_math_score':average_math_score.values})
avg_math_score_df

#Count of students passing Math by School
count_passing_math = passing_math_df.groupby('school_name')['math_score'].count()
count_passing_math

count_passing_math_df = pd.DataFrame({'school_name':count_passing_math.index, 'count_passing_math':count_passing_math.values})
count_passing_math_df

#Count of students passing Reading by School
count_passing_reading = passing_reading_df.groupby('school_name')['reading_score'].count()
count_passing_reading

count_passing_reading_df = pd.DataFrame({'school_name':count_passing_reading.index, 'count_passing_reading':count_passing_reading.values})
count_passing_reading_df

# Joining Schools with Average Reading Score 
schools_join_avgread_df = pd.merge(schools_df, avg_reading_score_df, on="school_name", how="outer")
schools_join_avgread_df.head()

# Joining Schools with Average Math Score
schools_join_avgmath_df = pd.merge(schools_join_avgread_df, avg_math_score_df, on="school_name", how="outer")
schools_join_avgmath_df.head()

# Joining Schools with Count of Students Passing Reading by School
schools_join_countread_df = pd.merge(schools_join_avgmath_df, count_passing_reading_df, on="school_name", how="outer")
schools_join_countread_df.head()

# Joining Schools with Count of Students Passing Math by School
schools_join_countmath_df = pd.merge(schools_join_countread_df, count_passing_math_df, on="school_name", how="outer")
schools_join_countmath_df.head()

#Calculating Percent of Students Passing Math by School
schools_join_countmath_df['percent_passing_math'] = (schools_join_countmath_df["count_passing_math"]/ schools_join_countmath_df["size"]) * 100
schools_join_countmath_df

#Calculating Percent of Student Passing Reading by School
schools_join_countmath_df['percent_passing_reading'] = (schools_join_countmath_df["count_passing_reading"]/ schools_join_countmath_df["size"]) * 100
schools_join_countmath_df

#Calculating Overall Passing Rate by School
schools_join_countmath_df['Overall_passing_rate'] = (schools_join_countmath_df["percent_passing_reading"] + schools_join_countmath_df["percent_passing_math"])/ 2
schools_join_countmath_df.head()

#Calculating Budget per Student by School
schools_join_countmath_df['Per_student_budget'] = (schools_join_countmath_df["budget"] / schools_join_countmath_df["size"])
schools_join_countmath_df.head()

#Renaming columns in dataframe
school_rename_df = schools_join_countmath_df.rename(columns={"school_name":"School Name", 
                                                             "type":"School Type",
                                                             "size": "Total Students", 
                                                             "budget": "Total School Budget", 
                                                             "Per_student_budget": "Per Student Budget", 
                                                             "avg_math_score": "Average Math Score",
                                                             "avg_reading_score": "Average Reading Score",
                                                             "percent_passing_math": "% Passing Math",
                                                             "percent_passing_reading": "% Passing Reading",
                                                             "Overall_passing_rate": "% Overall Passing Rate"})

school_rename_df.head()

#Reformating columns in dataframe
school_rename_df["Total School Budget"] = school_rename_df["Total School Budget"].map("${:,.2f}".format)
school_rename_df["Per Student Budget"] = school_rename_df["Per Student Budget"].map("${:,.2f}".format)
school_rename_df.head()

#Creating final dataframe
school_summary_df = school_rename_df.loc[:, ['School Name', 'School Type', 'Total Students',
       'Total School Budget', 'Per Student Budget','Average Reading Score', 'Average Math Score',
       '% Passing Math','% Passing Reading', '% Overall Passing Rate']]

school_summary_df.head()
```
