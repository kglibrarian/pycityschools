District Summary

```python
# High-level snapshot of districts key metrics

#Total Schools
schools_total = schools_df["school_name"].count()
schools_total

#Total Students
students_total = students_df["Student ID"].count()
students_total

#Total Budget
budget_total = schools_df["budget"].sum()
budget_total

#Average Math Score
math_average_total = students_df["math_score"].mean()
math_average_total_round = round(math_average_total, 6)
math_average_total_round

#Average Reading Score
reading_average_total = students_df["reading_score"].mean()
reading_average_total_round = round(reading_average_total, 6)
reading_average_total_round

# % Passing Math
passing_math_total = (passing_math_2["Student ID"].count()/ students_total) * 100
passing_math_total_round = round(passing_math_total, 6)
passing_math_total_round

# % Passing Reading
passing_reading_total = (passing_reading_2["Student ID"].count()/ students_total) * 100
passing_reading_total_round = round(passing_reading_total, 6)
passing_reading_total_round

# Overall Passing Rate
overall_passing_rate = (passing_reading_total_round + passing_math_total_round)/2
overall_passing_rate

# Creating a summary DataFrame using above values
high_level_snapshot_df = pd.DataFrame({"Total Schools": [schools_total],
                           "Total Students": [students_total],
                           "Total Budget": [budget_total],
                           "Average Math Score": [math_average_total_round],
                           "Average Reading Score": [reading_average_total_round],
                           "% Passing Math": [passing_math_total_round],
                           "% Passing Reading": [passing_reading_total_round],
                           "Overall Passing Rate": [overall_passing_rate]})

high_level_snapshot_df
```
